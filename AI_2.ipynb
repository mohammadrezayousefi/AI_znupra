```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
```

```python
xdata1 = np.random.rand(5, 1) * 10 -4
ydata1 = 4 * xdata1 + 5
# adding noise
ydata1 = ydata1 + np.random.normal(scale=1, size=ydata1.shape)
plt.plot(xdata1,ydata1,'ro')
```

```python
poly_features = PolynomialFeatures(degree=1, include_bias=False)
xdata1_poly = poly_features.fit_transform(xdata1)
regd1de1 = LinearRegression()
regd1de1.fit(xdata1_poly, ydata1)
```

```python
xdata1_vals = xdata1
xdata1_vals_poly = poly_features.transform(xdata1_vals)
ydata1_vals = regd1de1.predict(xdata1_vals_poly)
plt.suptitle('DATA 1 Degree 1')
plt.ylim((-20,20))
plt.xlim((-4,6))
plt.scatter(xdata1, ydata1)
plt.plot(xdata1_vals, y1_vals, color="gray")
plt.show()
```

```python
from sklearn import metrics
from sklearn.metrics import r2_score
print(f" intercept : {regd1de1.intercept_} | coefficient : {regd1de1.coef_} | MSE : {metrics.mean_squared_error(ydata1, ydata1_vals)} | R2 Score : {r2_score(ydata1, ydata1_vals)}")
```

```python
poly_features4 = PolynomialFeatures(degree=4, include_bias=False)
xdata1_poly4 = poly_features4.fit_transform(xdata1)
regd1de14 = LinearRegression()
regd1de14.fit(xdata1_poly4, ydata1)
```

```python
x1_vals_4_show = np.linspace(-5,5,500).reshape(-1,1)
x1_vals_poly_4_show = poly_features4.transform(x1_vals_4_show)
y1_vals_4_show = regd1de14.predict(x1_vals_poly_4_show)
x1_vals_4 = xdata1
x1_vals_poly_4 = poly_features4.transform(x1_vals_4)
y1_vals_4 = regd1de14.predict(x1_vals_poly_4)
plt.suptitle('DATA 1 : Degree 4')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(xdata1, ydata1)
plt.plot(x1_vals_4_show, y1_vals_4_show, color="r")
plt.show()
```

```python
print(f" intercept : {regd1de14.intercept_} | coefficient : {regd1de14.coef_} | MSE : {metrics.mean_squared_error(ydata1, y1_vals_4)} | R2 Score : {r2_score(ydata1, y1_vals_4)}")
```

```python

poly_features_16 = PolynomialFeatures(degree=16, include_bias=False)
x1_poly_16 = poly_features_16.fit_transform(xdata1)
reg1_1_16 = LinearRegression()
reg1_1_16.fit(x1_poly_16, ydata1)
x1_vals_16_show = np.linspace(-5,5,100).reshape(-1,1)
x1_vals_poly_16_show = poly_features_16.transform(x1_vals_16_show)
y1_vals_16_show = reg1_1_16.predict(x1_vals_poly_16_show)
x1_vals_16 = xdata1
x1_vals_poly_16 = poly_features_16.transform(x1_vals_16)
y1_vals_16 = reg1_1_16.predict(x1_vals_poly_16)
plt.suptitle('DATA 1 Degree 16')
plt.ylabel('Output')
plt.xlabel('Input')
plt.ylim((-20,20))
plt.xlim((-5,5))
plt.scatter(xdata1, ydata1)
plt.plot(x1_vals_16_show, y1_vals_16_show, color="gray")
plt.show()
```
